### Calculate the maximum number of items per article:
- sentences
- tokens
- events

```python
max_sent_cnt = 0
max_doc_len = 0
max_num_evt = 0
for evt in train_events:
    max_sent_cnt = len(evt['content']) if max_sent_cnt < len(evt['content']) else max_sent_cnt
    max_num_evt = len(evt['events']) if max_num_evt < len(evt['events']) else max_num_evt
    tmp_doc_len = 0
    for sent in evt['content']:
        tmp_doc_len += len(sent['tokens'])
    max_doc_len = tmp_doc_len if max_doc_len < tmp_doc_len else max_doc_len

print('max_sent_cnt', max_sent_cnt)
print('max_doc_len', max_doc_len)
print('max_num_evt', max_num_evt)
```

### Examine specific event samples

```python
list_of_interest = ['Hostile_encounter']
sample_count = 0
total_sentence_count = 0
for evt in train_events[:]:
    found = False
    found_cnt = 0
    sent_idx = []
    for typ in evt['events']:
        if typ['type'] in list_of_interest:
            found = True
            for ment in typ['mention']:
                sent_idx.append(ment['sent_id'])
            found_cnt+=1
            total_sentence_count+=1
    if found:
        sample_count+=1
        #for typ in evt['events']:
            #print(typ)#['type'])
        for idx in sent_idx:
            print(evt['content'][idx]['sentence'])
            print('--------------------------------')
        print('found_cnt',found_cnt)
        print('==================================================')
```
